```python
#Define a dict that will replace certain words in an effort to shorten column names for putting data into MySQL
rep = {'Latino':'Lat',
       'Hispanic':'His',
       'African_American':'AfAmer',
       'American_Indian':'AmInd',
       'Pacific_Islander':'PacIs',
       'Median':'Med',
       'Total':'Tot',
       'Alaska_Native':'AKNat',
       'Native_Hawaiian':'NatHI',
       'Family':'Fam',
       'households':'HH',
       'Male':'M',
       'Female':'F',
       'Institutionalized':'Inst',
       'population':'pop',
       'Correctional':'Corr',
       'facilities':'fac',
       'Noninstitutionalized':'NonInst'
       }
```

```python
def replaceAll(text,dic):
    """Function to replace text in a string using key:value pairs in a dict"""
    for i, j in dic.iteritems():
        text = text.replace(i, j)
    return text

#Apply replaceAll function to each element in a list
censusColEdit = [replaceAll(w,rep) for w in censusCol]
```

```python
#Replace column names with edited names
census.columns = censusColEdit
```
