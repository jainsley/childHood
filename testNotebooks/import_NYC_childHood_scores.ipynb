```python
#Function to read in school geojson data and insert it into a mongoDB database
def import_geojson_mongoDB(geo_file, coll):
    
    """This function will read in data in the form of geosjon and then insert it into a specified mongoDB database."""
    
    #Read in geojson data
    with open(geo_file) as f:
        json_file = json.load(f)
        
    #Import geojson data to specified collection
    for i in json_file['features']:
        db[coll].insert(i)
```

```python
def calc_NYCscore(Ser):
    from collections import Counter
    from numpy import nan
    entryCount = Counter()
    for entry in [Ser['school']['schoolScore'], Ser['crimeNYC'], Ser['censusNYC'], Ser['traffic']]:
        entryCount[entry] += 1
    if entryCount['No data.'] > 1:
        outputNYC = 'Insufficient data for this address.'
    elif Ser['school']['schoolScore'] == 'No data.':
        outputNYC = round((0.5 * (100 - Ser['crimeNYC'])) + (0.3 * Ser['censusNYC']) + (0.2 * (100 - Ser['traffic'])))
    elif Ser['crimeNYC'] == 'No data.':
        outputNYC = round((0.5 * float(Ser['school']['schoolScore'])) + (0.3 * Ser['censusNYC']) + (0.2 * (100 - Ser['traffic'])))
    elif Ser['censusNYC'] == 'No data.':
        outputNYC = round((0.5 * float(Ser['school']['schoolScore'])) + (0.35 * (100 - Ser['crimeNYC'])) + (0.15 * (100 - Ser['traffic'])))
    elif Ser['traffic'] == 'No data.':
        outputNYC = round((0.45 * float(Ser['school']['schoolScore'])) + (0.325 * (100 - Ser['crimeNYC'])) + (0.225 * Ser['censusNYC']))
    else:
        outputNYC = round((0.4 * float(Ser['school']['schoolScore'])) + (0.3 * (100 - Ser['crimeNYC'])) + (0.2 * Ser['censusNYC']) + (0.1 * (100 - Ser['traffic'])))
    return outputNYC
```

```python
CHdataframe['NYCscoreRecalc'] = CHdataframe.apply(calc_NYCscore, axis=1)
```

```python
import numpy as np

CHfixed = CHdataframe.replace('Insufficient data for this address.', np.nan)
CHsubset = CHfixed[['neighborhood', 'NYCscoreRecalc']]
```

```python
CHgrouped = CHsubset.groupby('neighborhood')
groupedMean = CHgrouped['NYCscoreRecalc'].aggregate(np.mean)
```

```python
#groupedSums.reset_index()
#CHgrouped.transform(np.mean)
hoodMean = groupedMean.reset_index()
hoodMean['zscore'] = (hoodMean['NYCscoreRecalc'] - hoodMean['NYCscoreRecalc'].mean())/hoodMean['NYCscoreRecalc'].std(ddof=0)
```

```python
hoodMean['scaledScore'] = (hoodMean['NYCscoreRecalc'] / hoodMean['NYCscoreRecalc'].max()) * 100
```
